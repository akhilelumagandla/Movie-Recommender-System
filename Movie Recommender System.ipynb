```python
import numpy as np
import pandas as pd
```

1. Importing DataSets

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
credits.head(2)['crew']
```

```python
# merge the two dataframes

movies = movies.merge(credits, on = 'title')
```

```python
movies.head(1)
```


2. Data Preprocessing

```python
movies.info()
```

```python
# genres
# id -> to fetch the movie posters
# keywords
# title
# overview
# cast
# crew

movies = movies[['genres', 'id', 'title', 'keywords','overview', 'cast', 'crew']]
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace = True)
```

```python
movies.iloc[0].genres
```

```python
# ast converts gneres (list with string indices) to list with int indices
import ast

def convert_to_tags(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
    
```

```python
movies['genres'] = movies['genres'].apply(convert_to_tags)
```

```python
movies['keywords'] = movies['keywords'].apply(convert_to_tags)
```

```python
movies.head()
```

```python
def cast_to_tags(obj):
    L = []
    cnt = 0
    for i in ast.literal_eval(obj):
        if cnt < 3:
            L.append(i['name'])
            cnt += 1
        else:
            break
        
    return L
```

```python
movies['cast'] = movies['cast'].apply(cast_to_tags)
```

```python
movies.iloc[0].crew
```

```python
def crew_to_tags(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
    
```

```python
movies['crew'] = movies['crew'].apply(crew_to_tags)
```

```python
movies.head()
```

```python
movies['overview'][0]
```

```python
# converting overview from str to a list

movies['overview'] = movies['overview'].apply(lambda x : x.split())
```

```python
movies['overview']
```

```python
movies.head()
```

```python
# now, for the model to work better
# we need to remove the spaces in between the words

movies['genres'].apply(lambda x : [i.replace(" ", "") for i in x])
```

```python
movies['genres'] = movies['genres'].apply(lambda x : [i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x : [i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x : [i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x : [i.replace(" ", "") for i in x])
```

```python
movies.head(2)
```

```python
# append all of other cols to tags

movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head(2)
```

```python
# making a new df only by taking the required columns

new_df = movies[['id', 'title', 'tags']]
```

```python
new_df.head()
```

```python
# convert tags to a string

new_df['tags'] = [" ".join(x) for x in new_df['tags']]
```

```python
new_df.info()
```

```python
new_df['tags'][0]
```

```python
# convert all the letters to lower case

new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df['tags'][0]
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
        
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

2. Vectorization

```python
# Bag of words technique

from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
# convert it to a numpy array

vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

```python
new_df['tags'][0]
```

3. Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
sorted(list(enumerate(similarity[0])), reverse = True, key = lambda x : x[1])
```

```python
def recommend(movie):
    movie_ind = new_df[new_df['title'] == movie].index[0]
    dist = similarity[movie_ind]
    L = sorted(list(enumerate(dist)), reverse = True, key = lambda x : x[1])[1 : 6]

    for i in L:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Inception')
```
